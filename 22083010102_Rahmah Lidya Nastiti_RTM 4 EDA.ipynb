Data covid-19

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data_covid = pd.read_csv('covid_19_indonesia_time_series_all.csv')
data_covid
```

Visualisasi data selama rentang beberapa tahun

```python

data_covid['Date'] = pd.to_datetime(data_covid['Date'])
data_covid['Year'] = data_covid['Date'].dt.to_period('Y')
covid_year = data_covid.groupby('Year').sum().reset_index()
covid_year['Year'] = covid_year['Year'].astype(str)
covid_year
```

visualisasi data

```python
plt.figure(figsize=(12,6))
plt.plot(covid_year['Year'],covid_year['New Cases'], color='skyblue', linewidth=2, label='Yearly Gruoped Time Series')
plt.xlabel('Year')
plt.ylabel('New Cases')
plt.legend()
plt.title('Yearly Gruped Time Series')
plt.show()
```

Pengelompokan (grouping) data time series berdasarkan bulan

```python

data_covid['Date'] = pd.to_datetime(data_covid['Date'])
data_covid['Month'] = data_covid['Date'].dt.to_period('M')
covid_month = data_covid.groupby('Month').sum().reset_index()
covid_month['Month'] = covid_month['Month'].astype(str)
covid_month
```

Pengelompokan (grouping) data time series berdasarkan hari

```python

data_covid['Date'] = pd.to_datetime(data_covid['Date'])
data_covid['Daily'] = data_covid['Date'].dt.to_period('D')
covid_daily = data_covid.groupby('Daily').sum().reset_index()
covid_daily['Daily'] = covid_daily['Daily'].astype(str)
covid_daily
```

Visualisasi bersama rata-rata mingguan kasus COVID-19 dan harian

```python

data_covid['Date'] = pd.to_datetime(data_covid['Date'])
data_covid['Weekly'] = data_covid['Date'].dt.to_period('W')
covid_weekly = data_covid.groupby('Weekly').mean().reset_index()
covid_weekly['Weekly'] = covid_weekly['Weekly'].astype(str)
covid_weekly
```

```python
plt.figure(figsize=(20,10))
plt.plot(covid_daily['Daily'],covid_daily['New Cases'], label='Kasus Harian')
plt.plot(covid_weekly['New Cases'], label='Rata-rata Mingguan')
plt.xlabel('Kasus COVID-19')
plt.ylabel('Tanggal')
plt.legend()
plt.show()
```

Hypothesis Testing rata-rata dari kasus harian

```python
from scipy.stats import ttest_1samp

mean_daily_cases = covid_daily['New Cases'].mean()
print('Rata-rata dari kasus harian:', mean_daily_cases)

t_stat, p_value = ttest_1samp(covid_daily['New Cases'], mean_daily_cases)  # Koreksi di sini
print('t_stat:', t_stat)
print('p_value:', p_value)
                              
alpha = 0.05

if p_value < alpha:  
    print('H0 ditolak, rata-rata dari kasus harian COVID-19 lebih dari', mean_daily_cases)
else:
    print('H0 diterima, rata-rata dari kasus harian COVID-19 sama dengan', mean_daily_cases)
```

Analisis korelasi

```python
x = data_covid['New Cases']
y = data_covid['New Deaths']
n = len(x)

sum_xy = np.sum(x * y)
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_x_squard = np.sum(x**2)
sum_y_squard = np.sum(y**2)
r = (n * sum_xy - sum_x * sum_y) / (np.sqrt(n * sum_x_squard - sum_x**2) * np.sqrt(n * sum_y_squard - sum_y**2))

print('koefisien korelasi :', r)
```

Model regresi 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data_covid = pd.read_csv('covid_19_indonesia_time_series_all.csv')

x = data_covid['New Cases']
y = data_covid['New Deaths']

def calculate_slope_intercept(x, y):
    n = len(x)
    
    totalx = np.sum(x)
    print('totalx:', totalx)
    totaly = np.sum(y)
    print('totaly:', totaly)
    
    xy = np.sum(x * y)
    print('totalxy:', xy)
    
    x_squared = np.sum(x**2)
    print('totalx2:', x_squared)
    
    a = ((n * xy) - (totalx * totaly)) / ((n * x_squared) - totalx**2)
    print('a =', a)
    
    b = ((n * totaly) - (totalx * xy)) / ((n * x_squared) - totalx**2)
    print('b =', b)
    
    y_pred = a * x + b
    print('Predicted Deaths:', y_pred)

calculate_slope_intercept(x, y)
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

data_covid = pd.read_csv('covid_19_indonesia_time_series_all.csv')

x = data_covid['New Cases'].values.reshape(-1, 1)
y = data_covid['New Deaths'].values

# Initialize and train the regressor
regressor = LinearRegression()
regressor.fit(x, y)

# Access the intercept and coefficients
intercept = regressor.intercept_
slope = regressor.coef_[0]

print('Intercept:', intercept)
print('Slope:', slope)
```

```python
regressor.intercept_
```

```python
y_pred = regressor.predict(x)
print(y_pred)
```

Evaluasi model regresi menggunakan MAPE

```python
def mape(y, y_pred):
  y, y_pred = np.array(y), np.array(y_pred)
  mape = np.mean(np.abs((y - y_pred) / y)) * 100
  return mape

mape_result = mape(y, y_pred)
print("MAPE:", mape_result)
```
